```python
df = df[~df['State'].isin(['Total (All India)', 'Total (States)', 'Total (Uts)'])]
print(df.head())
```

```python
print(df['State'].unique())
```

```python
statewise_df = df.drop(df[(df.State == 'Total (Uts)') | (df.State == 'Total (All India)') | 
               (df.State == 'Total (States)')].index)
```

```python
statewise_df.replace('Bankruptcy or Sudden change in Economic', 
           'Bankruptcy or Sudden change in Economic Status', inplace=True)
statewise_df.replace('By Other means (please specify)', 'By Other means', inplace=True)
statewise_df.replace('Not having Children(Barrenness/Impotency',
           'Not having Children (Barrenness/Impotency)', inplace=True)
```

```python
statewise_df = statewise_df.reset_index(drop=True)
statewise_df
```

```python
import seaborn as sns  #https://seaborn.pydata.org/generated/seaborn.set_style.html
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
#We use rc i.e a runtime configuration (rc) containing the default styles for every plot element you create
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
yearwise_total_suicide = statewise_df.groupby("Year")["Total"].sum()
print('Total number of suicides in each year: ', yearwise_total_suicide)
```

```python
# Plotting the yearwise total suicides in a line graph
yearwise_total_suicide.plot(kind="line", figsize=(9, 5), color='r')
plt.title('Yearwise Total Suicides')
plt.xlabel('Year')
plt.ylabel('Number of suicides')
plt.show()
```

```python
year_with_highest_suicides = yearwise_total_suicide.idxmax()
highest_suicides = yearwise_total_suicide.max()
print(f"The highest number of suicides occurred in the year {year_with_highest_suicides} with {highest_suicides} suicides.")
```

```python
statewise_total_suicide = statewise_df.groupby("State")["Total"].sum()
statewise_total_suicide
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x=statewise_total_suicide.index, y=statewise_total_suicide.values)
plt.xticks(rotation=90)
plt.title('Total Suicides by State')
plt.xlabel('State')
plt.ylabel('Total Suicides')
plt.show()
```

```python
#Creating a series of total suicides on the basis of states
statewise_total_suicide = statewise_df.groupby("State")["Total"].sum()
print('Total number of suicides in each State : ', statewise_total_suicide)

#plotting the above data
filter_state = pd.DataFrame(statewise_total_suicide.reset_index()) 
sort_state = filter_state.sort_values('Total',ascending = True)
f, ax = plt.subplots(figsize=(15, 4))
#sns.barplot(y = 'Total', x = 'State',data = sort_state,edgecolor = '0.5')
sns.barplot(y = 'Total', x = 'State',data = sort_state)
plt.xticks(rotation=75);
plt.show()
```

```python
causes_yearwise_suicide = statewise_df.groupby(['Year', 'Type'])['Total'].sum().unstack()
print(causes_yearwise_suicide)
```

```python
genderwise_suicide = statewise_df.groupby("Gender")["Total"].sum()
print('Total number of suicides by gender: ', genderwise_suicide)

# Plotting the genderwise suicides in a pie chart
plt.figure(figsize=(8, 5))
plt.pie(genderwise_suicide, labels=genderwise_suicide.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Suicides by Gender')
plt.show()
```

```python
# Plotting the suicide causes with year in a bar graph
causes_yearwise_suicide.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title('Yearwise Suicide Causes')
plt.xlabel('Year')
plt.ylabel('Number of Suicides')
plt.legend(title='Causes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#filtering the data as per the educational background
education_status = pd.DataFrame(statewise_df[statewise_df["Type_code"]=="Education_Status"].groupby(["Type","Gender"])["Total"].sum()).reset_index()
education_status

#plotting the above data
g = sns.catplot(y="Type", x="Total",hue="Gender", kind="bar", data=education_status,height=7, aspect=11.7/7);
sns.set_style('whitegrid');
```

```python
#filtering data according to the social status
filter_social_status = pd.DataFrame(statewise_df[statewise_df["Type_code"]=="Social_Status"].groupby(["Type","Gender"])["Total"].sum()).reset_index()

#plotting the above data
sns.catplot(x="Type", y="Total",hue="Gender", kind="bar", data=filter_social_status,height=7, aspect=11.7/7);
```

```python
#Age Distributiion
statewise_Age_df = statewise_df.drop(statewise_df[statewise_df.Age_group == '0-100+'].index)

counts = statewise_Age_df['Age_group'].value_counts().sort_index()
print(counts)

counts.plot(kind='bar', title='Age Distribution',figsize=(7,5))
plt.xticks(rotation=0)
plt.show()
```

```python
# Filtering the data for professional profiles
professional_df = statewise_df[statewise_df["Type_code"] == "Professional_Profile"]

# Grouping by profession type and summing the total suicides
professional_suicides = professional_df.groupby("Type")["Total"].sum()

# Finding the profession with the highest number of suicides
max_profession = professional_suicides.idxmax()
max_suicides = professional_suicides.max()

print(f"The professional background with the highest number of suicides is {max_profession} with {max_suicides} suicides.")
```

```python
filter_social_status = pd.DataFrame(statewise_df[statewise_df["Type_code"]=="Professional_Profile"].groupby(["Type","Gender"])["Total"].sum()).reset_index()
g = sns.catplot(x="Type", y="Total",hue="Gender", kind="bar", data=filter_social_status,height=7, aspect=11.7/7);
g.set_xticklabels(rotation=75);
```

```python
Based on the analysis of the dataset "Suicides in India 2001-2012", here are some observations and conclusions:

1. **Yearly Trends**:
    - The total number of suicides has generally increased from 2001 to 2011, with a peak in 2011 at 677,159 suicides.
    - There is a slight decline in the number of suicides in 2012 compared to 2011.

2. **State-wise Distribution**:
    - Maharashtra has the highest number of suicides among all states with 901,945 suicides.
    - West Bengal and Andhra Pradesh also have significantly high numbers of suicides, with 849,936 and 814,059 suicides respectively.
    - Lakshadweep has the lowest number of suicides with only 50 reported cases.

3. **Gender-wise Distribution**:
    - Males have a significantly higher number of suicides (4,641,088) compared to females (2,606,922).

4. **Causes of Suicides**:
    - The dataset includes various causes of suicides such as "Family Problems", "Illness", "Drug Abuse/Addiction", "Failure in Examination", etc.
    - Family problems appear to be a major cause of suicides across the years.

5. **Data Quality**:
    - There are some inconsistencies in the dataset, such as different spellings or representations of the same cause (e.g., "Bankruptcy or Sudden change in Economic" and "Bankruptcy or Sudden change in Economic Status").
    - Some columns have missing values which need to be addressed for a more accurate analysis.

6. **Visualization Insights**:
    - The bar plots and line graphs provide a clear visual representation of the trends and distributions, making it easier to identify patterns and outliers.

These observations highlight the critical areas that need attention for suicide prevention, such as mental health support, family counseling, and addressing socio-economic issues. The data also underscores the importance of gender-specific interventions and state-specific policies to effectively reduce the number of suicides.
```
