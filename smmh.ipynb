```python
df['5. What type of organizations are you affiliated with?'] = df['5. What type of organizations are you affiliated with?'].fillna(0)
print(df['5. What type of organizations are you affiliated with?'].head())
```

```python
print(df.isnull().sum())
```

```python
# Create a pivot table to find the relationship between social media usage and occupational status
pivot_table = df.pivot_table(index='4. Occupation Status', columns='6. Do you use social media?', aggfunc='size', fill_value=0)

# Plot the pivot table
pivot_table.plot(kind='bar', stacked=True)
plt.xlabel('Occupation Status')
plt.ylabel('Count')
plt.title('Relationship between Social Media Usage and Occupational Status')
plt.show()
```

```python
# Create a pivot table to find the relationship between social media usage and relationship status
pivot_table_relationship = df.pivot_table(index='3. Relationship Status', columns='6. Do you use social media?', aggfunc='size', fill_value=0)

# Display the pivot table
print(pivot_table_relationship)

# Plot the pivot table
pivot_table_relationship.plot(kind='bar', stacked=True)
plt.xlabel('Relationship Status')
plt.ylabel('Count')
plt.title('Relationship between Social Media Usage and Relationship Status')
plt.show()
```

```python
# Ensure the column '5. What type of organizations are you affiliated with?' is of string type
df['5. What type of organizations are you affiliated with?'] = df['5. What type of organizations are you affiliated with?'].astype(str)

# Create a pivot table to find the relationship between relationship status and type of organizations
pivot_table_org = df.pivot_table(index='3. Relationship Status', columns='5. What type of organizations are you affiliated with?', aggfunc='size', fill_value=0)

# Display the pivot table
print(pivot_table_org)

# Plot the pivot table
pivot_table_org.plot(kind='bar', stacked=True)
plt.xlabel('Relationship Status')
plt.ylabel('Count')
plt.title('Relationship between Relationship Status and Type of Organizations')
plt.show()
```

```python
# Create a pivot table to find the relationship between age, gender, and commonly used social media platforms
pivot_table_social_media = df.pivot_table(index='1. What is your age?', columns=['2. Gender', '7. What social media platforms do you commonly use?'], aggfunc='size', fill_value=0)

# Display the pivot table
print(pivot_table_social_media)

# Plot the pivot table
pivot_table_social_media.plot(kind='bar', stacked=True, figsize=(15, 7))
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Relationship between Age, Gender, and Commonly Used Social Media Platforms')
plt.show()
```

```python
# Create a pivot table to find the relationship between gender and occupation
pivot_table_gender_occupation = df.pivot_table(index='2. Gender', columns='4. Occupation Status', aggfunc='size', fill_value=0)

# Display the pivot table
print(pivot_table_gender_occupation)

# Plot the pivot table
pivot_table_gender_occupation.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Relationship between Gender and Occupation Status')
plt.show()
```

```python
# Create a pivot table to find the relationship between relationship status and occupation
pivot_table_relationship_occupation = df.pivot_table(index='3. Relationship Status', columns='4. Occupation Status', aggfunc='size', fill_value=0)

# Display the pivot table
print(pivot_table_relationship_occupation)

# Plot the pivot table
pivot_table_relationship_occupation.plot(kind='bar', stacked=True)
plt.xlabel('Relationship Status')
plt.ylabel('Count')
plt.title('Relationship between Relationship Status and Occupation Status')
plt.show()
```

```python
Based on the pivot tables created and analyzed, we can draw the following conclusions:

1. **Relationship between Social Media Usage and Occupational Status**:
    - The majority of university students use social media, followed by salaried workers.
    - Retired individuals and school students have the least social media usage.

2. **Relationship between Social Media Usage and Relationship Status**:
    - Single individuals have the highest social media usage, followed by those in a relationship.
    - Married individuals also use social media frequently, while divorced individuals have the least usage.

3. **Relationship between Relationship Status and Type of Organizations**:
    - Single individuals are mostly affiliated with universities, followed by schools.
    - Married individuals are more likely to be affiliated with companies and private organizations.
    - Those in a relationship have a diverse affiliation with various types of organizations.

4. **Relationship between Age, Gender, and Commonly Used Social Media Platforms**:
    - Younger individuals (ages 18-25) are more active on social media platforms like Instagram, YouTube, and Snapchat.
    - Females tend to use a wider variety of social media platforms compared to males.
    - Males are more active on platforms like Discord and Reddit.

5. **Relationship between Gender and Occupation Status**:
    - Females are predominantly university students, followed by salaried workers.
    - Males are also mostly university students, but there is a significant number of salaried workers.
    - Non-binary individuals have a very small representation in the dataset.

6. **Relationship between Relationship Status and Occupation**:
    - Single individuals are mostly university students, followed by school students.
    - Married individuals are predominantly salaried workers.
    - Those in a relationship have a significant number of university students and salaried workers.

These conclusions provide insights into the demographics and social media usage patterns of the surveyed individuals.
```
